4. Describe the architecture and functioning of a Recurrent Natural Network (RNN).

A Recurrent Neural Network (RNN) is a type of neural network designed to process sequential data, where the output depends not only on the current input but also on previous inputs in the sequence.
Ideal for time series, natural language, speech, and sequential data.
Unlike traditional feedforward networks, RNNs have loops to maintain memory of past inputs.

RNN Architecture
Basic Components - Input Layer ,Hidden Layer with Recurrence , Output Layer,

Functioning of an RNN
Initialization , Sequence Processing ,Backpropagation Through Time (BPTT)

Variants to Address RNN Limitations
Vanishing/Exploding Gradients , LSTM (Long Short-Term Memory) , GRU (Gated Recurrent Unit)

An RNN is a neural network that remembers past inputs via hidden states, making it ideal for sequential data tasks. Its architecture is centered on a recurrent hidden layer that loops information over time, while outputs can be produced at each time step or at the end of the sequence. Advanced variants like LSTM and GRU overcome its memory limitations for long sequences.

5. What is the key difference between LSTM and GRU networks in NLP
applications?

The key difference between LSTM (Long Short-Term Memory) and GRU (Gated Recurrent Unit) networks lies in their internal gating mechanisms and complexity, which affects how they handle sequential data in NLP tasks. Here's a detailed comparison

1. LSTM (Long Short-Term Memory)
Structure - Three gates:
Input gate вЂ“ controls how much new information flows into the cell state.
Forget gate вЂ“ controls which past information is discarded from the cell state.
Output gate вЂ“ controls what information from the cell state is output at the current step.
Maintains a cell state that acts as long-term memory.

2. GRU (Gated Recurrent Unit)
Structure - Two gates only:
Update gate вЂ“ decides how much of the past information to keep.
Reset gate вЂ“ controls how to combine new input with past memory.
No separate cell state; hidden state carries all memory.

NLP Application Implications
Machine Translation: LSTM often preferred for long sentences; GRU works well for shorter sentences.
Text Generation / Language Modeling: LSTM may handle longer context better.
Sentiment Analysis / Classification: GRU often suffices and trains faster.
Real-time applications: GRU preferred for lower latency and efficiency.

```python
# Import TextBlob library
from textblob import TextBlob

# Paragraph of text
text = """
I had a great experience using the new mobile banking app.
The interface is intuitive, and customer support was quick to resolve my issue.
However, the app did crash once during a transaction, which was frustrating.
"""

# Create a TextBlob object
blob = TextBlob(text)

# Perform sentiment analysis
sentiment = blob.sentiment

# Print polarity and subjectivity
print("Polarity:", sentiment.polarity)
print("Subjectivity:", sentiment.subjectivity)
```

```python
# Import NLTK libraries
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Download required NLTK resources (only needed once)
nltk.download('punkt_tab')

# Sample paragraph
text = """
Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical.
"""

# Tokenize the text into words
tokens = word_tokenize(text)

# Print tokenized words
print("Tokens:")
print(tokens)
print("\nTotal number of tokens:", len(tokens))

# Compute frequency distribution
freq_dist = FreqDist(tokens)

# Print frequency distribution
print("\nFrequency Distribution:")
for word, frequency in freq_dist.items():
    print(f"{word}: {frequency}")
```

```python
# Import necessary libraries
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Dummy dataset
texts = [
    "I love this project",             # Positive
    "This is an amazing experience",   # Positive
    "I hate waiting in line",          # Negative
    "This is the worst service",       # Negative
    "Absolutely fantastic!"            # Positive
]
labels = [1, 1, 0, 0, 1]  # 1=Positive, 0=Negative

# -------------------------------
# 1. Tokenization
# -------------------------------
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

print("Tokenized Sequences:")
print(sequences)

# -------------------------------
# 2. Padding sequences
# -------------------------------
max_len = max(len(seq) for seq in sequences)  # Maximum sequence length
padded_sequences = pad_sequences(sequences, maxlen=max_len, padding='post')

print("\nPadded Sequences:")
print(padded_sequences)

# -------------------------------
# 3. Convert labels to numpy array
# -------------------------------
labels = np.array(labels)

# -------------------------------
# 4. Build LSTM model
# -------------------------------
vocab_size = len(tokenizer.word_index) + 1  # +1 for padding token
embedding_dim = 50

model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_len))
model.add(LSTM(64, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))  # Binary classification

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# -------------------------------
# 5. Train the model
# -------------------------------
model.fit(padded_sequences, labels, epochs=50, verbose=1)

# -------------------------------
# 6. Test prediction
# -------------------------------
test_text = ["I really enjoyed this service"]
test_seq = tokenizer.texts_to_sequences(test_text)
test_pad = pad_sequences(test_seq, maxlen=max_len, padding='post')

prediction = model.predict(test_pad)
print("\nPrediction (probability of positive):", prediction[0][0])
```

```python
# Import spaCy
import spacy

# Load the English model (small)
nlp = spacy.load("en_core_web_sm")

# Sample paragraph
text = """
Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
development of IndiaвЂ™s atomic energy program. He was the founding director of the Tata
Institute of Fundamental Research (TIFR) and was instrumental in establishing the
Atomic Energy Commission of India.
"""

# Process the text with spaCy NLP pipeline
doc = nlp(text)

# -------------------------------
# 1. Tokenization and Lemmatization
# -------------------------------
print("Tokens and their Lemmas:")
for token in doc:
    print(f"Token: {token.text}, Lemma: {token.lemma_}")

# -------------------------------
# 2. Named Entity Recognition (NER)
# -------------------------------
print("\nNamed Entities:")
for ent in doc.ents:
    print(f"Entity: {ent.text}, Label: {ent.label_}")
```
